```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('fmri')
df
```

```python
sns.lineplot(x='timepoint',y='signal',data=df,hue ='region',markers=True)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=df,hue ='region',markers=True,style='event')
plt.show()
```

### iris

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
df
```

```python
sns.boxplot(x='species',y='petal_length',data=df)
plt.show()
```

```python
sns.boxplot(x='petal_length',data=df,hue='species')
```

```python
sns.kdeplot(x='sepal_width',data=df,hue='species')
```

###Penguins

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('penguins')
df
```

```python
sns.violinplot(x='body_mass_g',y='species',data=df)
```

```python
sns.kdeplot(x='flipper_length_mm',data=df, hue = 'species')
```
